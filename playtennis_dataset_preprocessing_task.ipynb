```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load the dataset
file_path = "/content/play tennis data set.csv"
data_full = pd.read_csv(file_path)

# Drop the 'Day' column as it is unnecessary for analysis
data_full = data_full.drop(columns=['Day'])

# Initialize the LabelEncoder
le = LabelEncoder()

# Apply Label Encoding to each categorical column
for column in data_full.columns:
    data_full[column] = le.fit_transform(data_full[column])

# Display the processed entire dataset
print(data_full)

```
